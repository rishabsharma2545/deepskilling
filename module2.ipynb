## **Sorting Algorithms Hands On**

```python
# 1. BUBBLE SORT
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# 2. INSERTION SORT
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# 3. HEAP SORT
def heapify(arr, n, i):
    largest = i
    left, right = 2 * i + 1, 2 * i + 2
    if left < n and arr[left] > arr[largest]: largest = left
    if right < n and arr[right] > arr[largest]: largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
    return arr


# 4. QUICK SORT
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[-1]
    left = [x for x in arr[:-1] if x <= pivot]
    right = [x for x in arr[:-1] if x > pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)


# 5. MERGE SORT
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # Merge process
    res = []
    while left and right:
        res.append(left.pop(0) if left[0] < right[0] else right.pop(0))

    return res + left + right
    

print(f"\nBubble Sort")
print(f"\tBest  Scenario : {bubble_sort([1, 2, 3, 4, 5]) = }")
print(f"\tAvg   Scenario : {bubble_sort([3, 1, 5, 4, 2]) = }")
print(f"\tWorst Scenario : {bubble_sort([5, 4, 3, 2, 1]) = }")

print(f"\nInsertion Sort")
print(f"\tBest  Scenario : {insertion_sort([1, 2, 3, 4, 5]) = }")
print(f"\tAvg   Scenario : {insertion_sort([3, 1, 5, 4, 2]) = }" )
print(f"\tWorst Scenario : {insertion_sort([5, 4, 3, 2, 1]) = }" )

print(f"\nHeap Sort")
print(f"\tBest  Scenario : {heap_sort([1, 2, 3, 4, 5]) = }")
print(f"\tAvg   Scenario : {heap_sort([3, 1, 5, 4, 2]) = }")
print(f"\tWorst Scenario : {heap_sort([5, 4, 3, 2, 1]) = }")

print(f"\nQuick Sort")
print(f"\tBest  Scenario : {quick_sort([3, 1, 5, 2, 4]) = }")
print(f"\tAvg   Scenario : {quick_sort([3, 5, 1, 4, 2]) = }")
print(f"\tWorst Scenario : {quick_sort([1, 2, 3, 4, 5]) = }")

print(f"\nMerge Sort")
print(f"\tBest  Scenario : {merge_sort([1, 2, 3, 4, 5]) = }")
print(f"\tAvg   Scenario : {merge_sort([3, 1, 5, 4, 2]) = }")
print(f"\tWorst Scenario : {merge_sort([4, 2, 5, 1, 3]) = }")
```

## **Searching Algorithms Hands On**

```python
# 1. LINEAR SEARCH
def linear_search(arr, target):
    
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1 


# 2. BINARY SEARCH
def binary_search(arr, target):

    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1


print("\nLinear Search")
print(f"\tBest   Scenario : {linear_search([10, 20, 30, 40, 50], 10) = }")
print(f"\tAvg    Scenario : {linear_search([10, 20, 30, 40, 50], 30) = }")
print(f"\tWorst  Scenario : {linear_search([10, 20, 30, 40, 50], 99) = }")

print("\nBinary Search")
print(f"\tBest   Scenario : {binary_search([10, 20, 30, 40, 50, 60, 70], 40) = }")
print(f"\tAvg    Scenario : {binary_search([10, 20, 30, 40, 50, 60, 70], 20) = }")
print(f"\tWorst  Scenario : {binary_search([10, 20, 30, 40, 50, 60, 70], 99) = }")
```

## **Arrays Hands On**



```python
# 1. CUSTOM REPLICAS OF BUILT-IN FUNCTIONS

def custom_map(func, arr):

    res = []
    for item in arr:
        res.append(func(item))
    return res

def custom_filter(func, arr):

    res = []
    for item in arr:
        if func(item):
            res.append(item)
    return res

def custom_reduce(func, arr, initial):

    accumulator = initial
    for item in arr:
        accumulator = func(accumulator, item)
    return accumulator

def custom_enumerate(arr):
    
    res = []
    for i in range(len(arr)):
        res.append((i, arr[i]))
    return res

def kadanes_algorithm(arr):
    
    if not arr: return 0
    global_max = arr[0]
    current_max = arr[0]
    
    for x in arr[1:]:
        current_max = max(x, current_max + x)
        global_max = max(global_max, current_max)
    return global_max

def prefix_sum(arr):
    
    if not arr: return []
    res = [arr[0]]
    for i in range(1, len(arr)):
        res.append(res[-1] + arr[i])
    return res


double = lambda x: x * 2
is_even = lambda x: x % 2 == 0
add = lambda x, y: x + y

print("\nCustom Functional Algorithms")
print(f"\tCustom Map (Double)      : {custom_map(double, [1, 2, 3, 4, 5]) = }")
print(f"\tCustom Filter (Evens)    : {custom_filter(is_even, [1, 2, 3, 4, 5]) = }")
print(f"\tCustom Reduce (Sum)      : {custom_reduce(add, [1, 2, 3, 4, 5], 0) = }")
print(f"\tCustom Enumerate         : {custom_enumerate(['a', 'b', 'c']) = }")


print("\nKadane's Algorithm (Max Subarray Sum)")
print(f"\tBest   Scenario (All Pos) : {kadanes_algorithm([1, 2, 3, 4, 5]) = }")
print(f"\tAvg    Scenario (Mixed)   : {kadanes_algorithm([2, -1, 3, 4, -2, 1]) = }")
print(f"\tWorst  Scenario (All Neg) : {kadanes_algorithm([-5, -2, -3, -4, -6]) = }")

print("\nPrefix Sum Array")
print(f"\tBest   Scenario (Single)   : {prefix_sum([10]) = }")
print(f"\tAvg    Scenario (All Pos)  : {prefix_sum([1, 2, 3, 4, 5]) = }")
print(f"\tWorst  Scenario (Mixed)    : {prefix_sum([10, -5, 20, -10]) = }")
```
