**Project Creation Step**  
* django:  
    > django startproject _**projectname**_

* flask:  
    > mkdir _**projectname**_    
    > create file _**app.py**_  
    > create file _**config.py**_  
    > create subfolder _**router**_ having _**init.py**_ & _**routes.py**_  

* fastapi: 
    > mkdir _**projectname**_  
    > create file _**main.py**_  
    > create file _**schemas.py**_
