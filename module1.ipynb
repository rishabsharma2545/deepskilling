# **SOLID Principles Notes**

* **S***RP : Single Responsibility Principle*
    > Only one responsibility and only one reason to change.

    **Benefits:**
    - Testing – A class with one responsibility will have far fewer test cases.
    - Lower coupling – Less functionality in a single class will have fewer dependencies.
    - Organization – Smaller, well-organized classes are easier to search than monolithic ones.

---

* **O***CP : Open / Closed Principle*
    > Open for extension but Closed for modification to avoid causing potential new bugs.

    **Benefits:**
    - Reduced risk of introducing bugs in existing, tested code.
    - Easier to add new features without changing current implementation.
    - Improves maintainability and scalability of applications.

---

* **L***SP : Liskov Substitution Principle*
    > If A -> B, then we can replace A with B without disrupting the behavior of our program.

    **Benefits:**
    - Ensures subclasses remain compatible with parent classes.
    - Promotes reliable inheritance and polymorphism.
    - Makes code easier to extend and refactor safely.

---

* **I***SP : Interface Segregation Principle*
    > Larger interfaces should be split into smaller ones with limited methods.

    **Benefits:**
    - Prevents classes from implementing unnecessary methods.
    - Reduces the impact of changes by limiting dependencies.
    - Leads to cleaner, more focused, and maintainable code.

---

* **D***IP : Dependency Inversion Principle*
    > High-level modules should not depend on low-level modules; both depend on abstractions.

    **Benefits:**
    - Reduces tight coupling between components.
    - Makes unit testing easier through mocking and dependency injection.
    - Improves flexibility by allowing swapping of implementations without affecting business logic.

---

# **Design Patterns Notes**

> ## _Creational Patterns:_
> Focus on object creation mechanisms.

### **Singleton Pattern**
- Ensures that only one instance of a class exists.
- Provides a global access point to that instance.
- Useful for shared resources such as logging, configuration, caching, and database connections.

**Advantages:**
- Controlled access to a single instance.
- Reduces memory usage.
- Ensures consistency across the application.

---

### **Factory Pattern**
- Creates objects without exposing the object creation logic.
- Delegates object instantiation to subclasses or factory methods.
- Useful when object creation is complex or depends on runtime conditions.

**Advantages:**
- Promotes loose coupling.
- Simplifies object creation.
- Makes code easier to extend and maintain.

---

### **Builder Pattern**
- Constructs complex objects step by step.
- Separates object construction from representation.
- Allows creating different variations of the same object.

**Advantages:**
- Improves readability of object creation.
- Handles complex constructors effectively.
- Provides flexibility in object configuration.

---

> ## _Structural Patterns:_
> Focus on how classes and objects are composed.

### **Adapter Pattern**
- Makes incompatible interfaces work together.
- Acts as a bridge between two incompatible systems.
- Converts one interface into another expected by the client.

**Advantages:**
- Improves compatibility between existing components.
- Encourages code reuse.
- Reduces modification of existing code.

---

### **Decorator Pattern**
- Dynamically adds new behavior to an object.
- Provides an alternative to inheritance for extending functionality.
- Wraps objects with additional responsibilities at runtime.

**Advantages:**
- Flexible feature addition.
- Avoids large inheritance hierarchies.
- Supports Open/Closed Principle.

---

### **Proxy Pattern**
- Controls access to another object.
- Acts as a placeholder or surrogate for a real object.
- Can perform access control, caching, logging, or lazy initialization.

**Advantages:**
- Enhances security and control.
- Improves performance through caching.
- Supports lazy loading of resources.

---

### **Facade Pattern**
- Provides a simplified interface to a complex subsystem.
- Hides implementation details from clients.
- Offers a single entry point to multiple components.

**Advantages:**
- Reduces system complexity.
- Improves code readability.
- Promotes loose coupling between clients and subsystems.

---

> ## _Behavioral Patterns:_
> Focus on communication and interaction between objects.

### **Observer Pattern**
- Defines a one-to-many dependency between objects.
- When one object changes state, all dependent objects are notified automatically.
- Commonly used in event-driven systems.

**Advantages:**
- Promotes loose coupling.
- Supports event handling and notifications.
- Simplifies communication between components.

**Examples:**
- YouTube subscriptions
- Email notifications
- Stock market alerts

---

### **Strategy Pattern**
- Defines a family of interchangeable algorithms.
- Encapsulates each algorithm in a separate class.
- Allows selecting algorithms dynamically at runtime.

**Advantages:**
- Eliminates large conditional statements.
- Makes algorithms interchangeable.
- Improves maintainability and flexibility.

**Examples:**
- Payment methods (UPI, Credit Card, PayPal)
- Sorting strategies
- Compression algorithms

---

### **Command Pattern**
- Encapsulates a request as an object.
- Separates the sender from the receiver.
- Supports queuing, logging, and undo/redo operations.

**Advantages:**
- Decouples request sender and receiver.
- Enables command history and rollback.
- Improves extensibility.

**Examples:**
- Undo/Redo functionality
- Menu actions
- Task scheduling systems

---

> ## _Architectural Patterns:_
> Focus on the high-level structure of applications.

### **MVC (Model-View-Controller) Pattern**
Separates an application into three components:

#### Model
- Represents data and business logic.
- Interacts with databases and services.

#### View
- Represents the user interface.
- Displays data to users.

#### Controller
- Handles user input.
- Updates the Model and View accordingly.

**Advantages:**
- Separation of concerns.
- Easier maintenance and testing.
- Supports parallel development.

**Used In:**
- Django (MVT variation)
- Spring MVC
- ASP.NET MVC

---

### **MVVM (Model-View-ViewModel) Pattern**
An evolution of MVC that introduces a ViewModel between the View and Model.

#### Model
- Contains business logic and data.

#### View
- Displays information to users.

#### ViewModel
- Acts as an intermediary between View and Model.
- Exposes data and commands to the View.
- Supports data binding.

**Advantages:**
- Better separation of UI and business logic.
- Easier unit testing.
- Supports two-way data binding.

**Used In:**
- Angular
- WPF
- Xamarin
- .NET MAUI

---
